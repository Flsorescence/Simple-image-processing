```python
# img = cv.imread('D:\\mego.jpg')

def cb1(): 
    global img
    img = cv.imread('D:\\mego.jpg')
    cv.namedWindow("Image")
    cv.imshow("Image", img)
    cv.waitKey(0)
    cv.destroyAllWindows()
    
def cb2():  
    global img
    height = int(e1.get())
    width = int(e2.get()) 
    resize = cv.resize(img, (width, height), interpolation=cv.INTER_CUBIC)
    cv.imshow("resize", resize)
    print(height, width)

def rotate(image, angle, center=None, scale=1.0):
    (h, w) = image.shape[:2]
    if center is None:
        center = (w / 2, h / 2)
    M = cv.getRotationMatrix2D(center, angle, scale)
    rotated = cv.warpAffine(image, M, (w, h))
    return rotated    

def rotate1(image, angle):  
    # grab the dimensions of the image and then determine the  
    # center  
    (h, w) = image.shape[:2]  
    (cX, cY) = (w // 2, h // 2)  
  
    # grab the rotation matrix (applying the negative of the  
    # angle to rotate clockwise), then grab the sine and cosine  
    # (i.e., the rotation components of the matrix)  
    M = cv.getRotationMatrix2D((cX, cY), -angle, 1.0)  
    cos = np.abs(M[0, 0])  
    sin = np.abs(M[0, 1])  
  
    # compute the new bounding dimensions of the image  
    nW = int((h * sin) + (w * cos))  
    nH = int((h * cos) + (w * sin))  
  
    # adjust the rotation matrix to take into account translation  
    M[0, 2] += (nW / 2) - cX  
    M[1, 2] += (nH / 2) - cY  
  
    # perform the actual rotation and return the image  
    return cv.warpAffine(image, M, (nW, nH), cv.BORDER_REPLICATE )  

def cb3():  
    angle = int(e3.get())
    print(angle)
    rotated = rotate1(img, angle)  
    cv.imshow("rotated", rotated)
    
def cb4(v1):
    v = v1.get()
    dst = img
    if v == 0:
        x = cv.Sobel(img,cv.CV_16S,1,0)  
        y = cv.Sobel(img,cv.CV_16S,0,1) 
        absX = cv.convertScaleAbs(x)   # 转回uint8  
        absY = cv.convertScaleAbs(y)  
        dst = cv.addWeighted(absX,0.5,absY,0.5,0)
        name = "Sobel"
    else:
        gray_lap = cv.Laplacian(img,cv.CV_16S,ksize = 1)  
        dst = cv.convertScaleAbs(gray_lap)
        name = "Laplace"
    cv.imshow(name, dst)
        
def cb5():
    dst = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
    rows,cols = dst.shape
    dst1 = dst.copy()
    a = 1.2
    b = 50
    for i in range(rows):
        for j in range(cols):
            color = dst[i,j]*a+b
            if color < 64:
                dst1[i,j] = dst[i,j] / 2
            elif color < 192:
                dst1[i,j] = dst[i,j] / 2 + dst[i,j]
            else:
                dst1[i,j] = dst[i,j] / 2
    cv.imshow("Contrast", dst1)
    
def cb6():
    temp = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
    equ = cv.equalizeHist(temp) 
    cv.imshow("equalization", equ)
    
def cb7():
    smooth = cv.medianBlur(img,5)  
    cv.imshow("smooth", smooth)
    
def cb8():
    fushi = img
    kernel = np.ones((5,5),np.uint8) 
    erosion = cv.erode(fushi, kernel, iterations = 1)  
    cv.imshow("erosion", erosion)
    
def cb9():
    pengzhang = img
    kernel = np.ones((5,5),np.uint8) 
    dilation = cv.dilate(img,kernel,iterations = 1)
    cv.imshow("dilation", dilation)

def cb10():  
    boundary = int(e4.get())
    if boundary < 0:
        boundary = 0
    elif boundary > 255:
        boundary = 255
    dst = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
    ret,result = cv.threshold(dst, boundary, 255, cv.THRESH_BINARY)
    cv.imshow("segmentation", result)
    print(boundary) 
    
def cb11(v2): 
    v = v2.get()
    if v == 0:
        dst = cv.cvtColor(img, cv.COLOR_RGB2XYZ)
        name = "XYZ"
    elif v == 1:
        dst = cv.cvtColor(img, cv.COLOR_RGB2YCrCb)
        name = "YCrCb"
    else:
        dst = cv.cvtColor(img, cv.COLOR_RGB2HSV)
        name = "HSV"
    cv.imshow(name, dst)

    
root = tk.Tk()  
root.title("dispose-picture")
root.geometry('520x300')

f1 = tk.Frame(root,width=20,height=10,borderwidth=2)
f2 = tk.LabelFrame(root,width=20,height=10,padx=2, pady = 2)
f3 = tk.LabelFrame(root,width=20,height=10,padx=2, pady = 2)
f4 = tk.LabelFrame(root,width=20,height=10,padx=2, pady = 2)
f5 = tk.Frame(root,width=20,height=10,borderwidth=2)
f6 = tk.LabelFrame(root,width=20,height=10,padx=2, pady = 2)
f7 = tk.LabelFrame(root,width=20,height=10,padx=2, pady = 2)

f1.grid(row=0, column=0)
f2.grid(row=1, column=0)
f3.grid(row=2, column=0)
f4.grid(row=0, column=1)
f5.grid(row=1, column=1, rowspan=2)
f6.grid(row=0, column=2)
f7.grid(row=1, column=2)


b1 = tk.Button(f1, text = '打开图片',width = 20,height = 1, command = cb1)  
b1.pack()
l21 = tk.Label(f2, text = '高：', width = 5, height = 1)
l21.grid(row = 0, column = 0)
e1 = tk.Entry(f2, show = None)
e1.grid(row = 0, column = 1)
l22 = tk.Label(f2, text = '宽：', width = 5, height = 1)
l22.grid(row = 1, column = 0)
e2 = tk.Entry(f2, show = None)
e2.grid(row = 1, column = 1)
b2 = tk.Button(f2, text = '缩放', width = 20, height = 1, command = cb2)
b2.grid(row = 2, column = 0, columnspan=2)

l3 = tk.Label(f3, text = '旋转角度', width = 10, height = 1)
l3.pack()
e3 = tk.Entry(f3, show = None)
e3.pack()
b3 = tk.Button(f3, text = '旋转', width = 10, height = 1, command = cb3)
b3.pack()


LANGS1 = [
    ('Sobel',0),
    ('Laplace',1)]
v1 = tk.IntVar()
v1.set(0)
for lang,num in LANGS1:
    b41 = tk.Radiobutton(f4,text=lang,variable=v1,value=num)
    b41.pack()
b4 = tk.Button(f4, text = '锐化', width = 20, height = 1, command = lambda: cb4(v1))
b4.pack()

b5 = tk.Button(f5, text = '线性拉伸', width = 20, height = 1, command = cb5)
b5.pack()
b6 = tk.Button(f5, text = '直方图均衡', width = 20, height = 1, command = cb6)
b6.pack()
b7 = tk.Button(f5, text = '中值滤波', width = 20, height = 1, command = cb7)
b7.pack()
b8 = tk.Button(f5, text = '腐蚀', width = 20, height = 1, command = cb8)
b8.pack()
b9 = tk.Button(f5, text = '膨胀', width = 20, height = 1, command = cb9)
b9.pack()

l4 = tk.Label(f6, text = '阈值：', width = 10, height = 1)
l4.pack()
e4 = tk.Entry(f6, show = None)
e4.pack()
b10 = tk.Button(f6, text = '分割', width = 10, height = 1, command = cb10)
b10.pack()

LANGS2 = [
    ('XYZ',0),
    ('YCbCr',1),
    ('HSV',2)]
v2 = tk.IntVar()
v2.set(0)
for lang,num in LANGS2:
    b111 = tk.Radiobutton(f7,text=lang,variable=v2,value=num)
    b111.pack()
b11 = tk.Button(f7, text = '色彩空间转换', width = 20, height = 1, command=lambda :cb11(v2))
b11.pack()

root.mainloop()  
```
